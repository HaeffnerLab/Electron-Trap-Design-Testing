```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy as sp
from scipy import stats
from scipy import optimize
import pandas as pd
import scipy.optimize as fitter
%matplotlib inline
```

```python
# loading data
#data_background = np.loadtxt("000.asc", skiprows = 5)
#data_config1 = np.loadtxt("01.asc", skiprows = 5)
data_bg1 = np.loadtxt("background1.asc", skiprows=5) #background
data_bg2 = np.loadtxt("background2.asc", skiprows=5)
data1 = np.loadtxt("01.asc", skiprows=5)
data2 = np.loadtxt("partially_short.asc", skiprows=5)
data3 = np.loadtxt("two_short.asc", skiprows=5)
```

```python
# Plotting data 1
f = np.linspace(0, 3000, 250)
plt.figure(figsize = (10, 6))
#plt.plot(f, data1-data_bg, "b-", label="not shorted")
plt.plot(f, data2-data_bg1, "r-", label="partially shorted")
plt.plot(f, data3-data_bg2, "b-", label="shorted at two points")
plt.xlabel('f (MHz)')
plt.ylabel('Response (dB)')
plt.grid(True)
plt.legend()
```
