```python
# loading data
data_background = np.loadtxt("000.asc", skiprows = 5)
data_config1 = np.loadtxt("01.asc", skiprows = 5)
data_config2 = np.loadtxt("02.asc", skiprows = 5)
data_config3 = np.loadtxt("03.asc", skiprows = 5)
data_config4 = np.loadtxt("04.asc", skiprows = 5)

# data taken on June 20th
data_background2 = np.loadtxt("0620000.asc", skiprows = 5)
data_config5 = np.loadtxt("05.asc", skiprows = 5)
data_config6 = np.loadtxt("06.asc", skiprows = 5)
data_config7 = np.loadtxt("07.asc", skiprows = 5)
```

```python
# loading simulation

# input the configuration number, output the frequency and dB arrays from the 
# simulation csv file
def load_sim(config):
    sim = np.loadtxt("Bantam_test"+str(config)+".csv", skiprows=1, delimiter=",")
    sim_f = []
    sim_dB = []
    for sub in sim:
        sim_f.append(sub[0])
        sim_dB.append(sub[1])
    sim_f = np.array(sim_f) * 1000.0
    sim_dB = np.array(sim_dB)
    return (sim_f, sim_dB)

sim_f_1, sim_dB_1 = load_sim(1)
sim_f_2, sim_dB_2 = load_sim(2)
sim_f_4, sim_dB_4 = load_sim(4)

```

```python
# Plotting configuration 1 data and simulation
f = np.linspace(0, 3000, 250)
plt.figure(figsize = (10, 6))
plt.plot(f, data_config1 - data_background, "r.", label = "config 1 data")
plt.plot(sim_f_1, sim_dB_1, label = "config 1 sim")
#plt.plot(f, data_config3 - data_background, label = "config 3 data")
#plt.plot(f, data_config4 - data_background, label = "config 4 data")
plt.xlabel('f (MHz)')
plt.ylabel('Response (dB)')
plt.legend()
plt.grid(True)
```
