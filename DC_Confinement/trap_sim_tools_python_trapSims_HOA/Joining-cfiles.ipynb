```python
import pickle
import numpy as np

#Load:
files = [ open( 'BEMSolver_cfiles/atrap_cct_simulation_{}.pkl'.format(j) ) for j in range(1,8) ]
pkls = [ pickle.load(f) for f in files ]

#Concatenate along the z axis:
xlen = len(pkls[0].X)
ylen = len(pkls[0].Y)
pkl_output = pkls[0]
#Concatenate z's of electrodes:
for el in range(0,24):
    el_name = 'EL_DC_{}'.format(el)
    el_pot_tuple  = ( pkls[0][el_name],)
    #initial z is repeated on pickle files after pikls[0]:
    init_z  =  1
    el_pot_tuple  = el_pot_tuple + tuple( [ pkl[el_name][:,:, init_z:] for pkl in pkls[1:] ] )
    pkl_output[el_name] = np.concatenate( el_pot_tuple, axis=2 )

#Concatenate Z's:
#initial z is repeated on pickle files after pikls[0]:
Zs_tuple  = (pkls[0]['Z'],)
Zs_tuple  = Zs_tuple + tuple( [ pkl['Z'][1:] for pkl in pkls[1:] ] )
pkl_output['Z'] = np.concatenate( Zs_tuple )

#Save output:
f = open('data.pkl','w')
pickle.dump(pkl_output,f)

f = open('data.pkl')
pkl = pickle.load(f)
```
