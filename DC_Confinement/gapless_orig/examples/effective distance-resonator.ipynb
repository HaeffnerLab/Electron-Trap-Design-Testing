Now walk the height and find the optimum electrode size for each height

```python
def find_optimal_electrode_size(deff, electrode_size):
    # return a 2-tuple containing (optimal_electrode_size, effective_distance)
    j = where(deff==min(deff))
    return ( list(electrode_size[j])[0], min(deff) )
xp = 0.; yp = 0.
height = arange(10e-6, 140e-6, 5e-6)
electrode_size = arange(20e-6, 800e-6, 1e-6)

optimal_size = []
optimal_deff = []

for h in height:
    deff = []
    for d in electrode_size:
        w = World(2*pi*35*1e6, d=1e-7)
        w.add_electrode('cnt', (-1*d/2, d/2), (-1*d/2, d/2), 'dc', 1.0) # pickup electrode
        cnt = w.electrode_dict['cnt']
        deff.append(cnt.compute_d_effective(xp, yp, h)[2]*1e6)
    L, d = find_optimal_electrode_size(deff, electrode_size)
    optimal_size.append(L)
    optimal_deff.append(d)
```

Fit the calculated capacitance from the finite element solver

```python
elec_length = array([50., 80., 100., 120., 140., 170., 200.])*1e-6
cap = array([16.9, 21.5, 24.8, 28.35, 32.0, 37.7, 43.6])*1e-15 # femtofarad
c1, c0 =  polyfit(elec_length, cap, 1)
capacitance = lambda L: c0 + c1*L

def find_optimal_size(h, doPlot=False):
    electrode_size = arange(20e-6, 500e-6, 1e-6)
    coupling_coeff = [] # capacitance*d^2
    for d in electrode_size:
        w = World(2*pi*35*1e6, d=1e-7)
        w.add_electrode('cnt', (-1*d/2, d/2), (-1*d/2, d/2), 'dc', 1.0) # pickup electrode
        cnt = w.electrode_dict['cnt']
        deff = cnt.compute_d_effective(0., 0., h)[2]
        capn = capacitance(d)
        coupling_coeff.append( capn*(deff**2) )
    coupling_coeff = array(coupling_coeff)
    j = where(coupling_coeff==min(coupling_coeff))
    if doPlot:
        plot(electrode_size*1e6, coupling_coeff)
    return ( list(electrode_size[j])[0], min(coupling_coeff) )

q = 1.6e-19 # coulomb
f = 2*pi*1e6 # 1 MHz frequency
m = 6.7e-26 # kg
height = arange(30e-6, 140e-6, 5e-6)
g = []
size = []
for h in height:
    (s, c) = find_optimal_size(h)
    #print c
    g.append( (1/4.)*(q**2/(f*m*c) ) )
    size.append(s)
fig, ax = subplots(1,1)
ax.plot(height*1e6, g)
ax.set_xlabel('Ion height (um)')
ax.set_ylabel('Coupling energy (Hz)')
ax.set_title('Coupling energy given optimal pickup electrode size')
```
