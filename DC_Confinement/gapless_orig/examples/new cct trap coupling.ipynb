Explore the coupling with the fixed wire parameters we have

```python
from gapless import *
C = 4.9e-14
d = 120e-6
q = 1.6e-19 # coulomb
f = 2*pi*200e3 # 1 MHz frequency
m = 6.7e-26 # kg
xp = 0.; yp = 0.
height = arange(25e-6, 100e-6, 1e-6)
g = []
for h in height:
    w = World(2*pi*35*1e6, d=1e-7)
    w.add_electrode('cnt', (-1*d/2, d/2), (-1*d/2, d/2), 'dc', 1.0) # pickup electrode
    cnt = w.electrode_dict['cnt']
    deff = cnt.compute_d_effective(0., 0., h)[2]
    coupling_coeff = C*(deff**2)
    g.append((1/4.)*(q**2/(f*m*coupling_coeff)))
fig, ax = subplots(1,1)
ax.plot(height*1e6, g)
ax.set_xlabel('Ion height (um)')
ax.set_ylabel('Coupling energy (Hz)')
    
```
