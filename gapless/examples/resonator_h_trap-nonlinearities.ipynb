```python
import scipy.integrate as integrate
m = 40 * const.atomic_mass
q = const.elementary_charge
temp_list = logspace(-2,1,100)
freqlist = []
distlist = []
for temp in temp_list:
    p1 = p0
    p1[-1] = -temp * ktoev
    r0 = roots(p1)
    r0 = r0[~numpy.iscomplex(r0)]
    zstart = real(r0[2])
    zstop = real(r0[1])
    distlist.append(zstop-zstart)
    potfun = lambda z : polyval(p0,z) 
    fun = lambda z : 1/np.sqrt(abs(potfun(zstart)-potfun(z) + 1e-25))
    t = sqrt(m/(2*q)) * integrate.quad(fun, zstart, zstop)[0] * 2
    freqlist.append(1/t)
freqlist = array(freqlist)
plot(temp_list,1-freqlist/freqlist[0])
ylabel('Relative frequency change')
xlabel('Temperature (K)')
figure()
distlist = array(distlist)*1e6
plot(temp_list, distlist)
ylabel('Ion movement (um)')
xlabel('Temperature (K)')
print 'Frequency: ',max(freqlist)/1e6
```

```python
from scipy import optimize as opt
def find_trap_location(r0):
    x0 = [r0[0], r0[1]]
    fixed = r0[2]
    cost = lambda r: w.compute_pseudopot([r[0],r[1],fixed])
    result = opt.minimize(cost, x0, method='Nelder-Mead')
    r_trap = [result.x[0], result.x[1], fixed ]
    return array(r_trap)

r0 = array([-16, 102, 760])*1e-6
r_trap = find_trap_location(r0)
print r_trap*1e6
print w.compute_pseudopot(r_trap)
```
